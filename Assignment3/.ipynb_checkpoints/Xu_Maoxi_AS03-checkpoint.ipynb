```python
 # Maoxi Xu
 # Import necessary libraries
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import SGDClassifier, LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = load_iris()
XX = iris.data[:, :2]  # Extract only the Sepal Length and Sepal Width features
YY = iris.target
```

```python
 # Split the dataset into training and testing sets
random_state_xyz = 193  # Replace 193 with the last three digits of your student ID
XX_train, XX_test, YY_train, YY_test = train_test_split(XX, YY, test_size=0.20, random_state=random_state_xyz)
```

```python
 # Initialize the models
models = {
    "SGD (Hinge)": SGDClassifier(loss="hinge", random_state=random_state_xyz),
    "SGD (Modified Huber)": SGDClassifier(loss="modified_huber", random_state=random_state_xyz),
    "Logistic Regression": LogisticRegression(random_state=random_state_xyz),
    "SVM (Linear Kernel)": SVC(kernel="linear", random_state=random_state_xyz),
    "SVM (RBF Kernel)": SVC(kernel="rbf", random_state=random_state_xyz),
    "Decision Tree": DecisionTreeClassifier(criterion="gini", random_state=random_state_xyz),
    "Random Forest": RandomForestClassifier(random_state=random_state_xyz),
    "KNN (k=3)": KNeighborsClassifier(n_neighbors=3),
    "KNN (k=5)": KNeighborsClassifier(n_neighbors=5)
}
```

```python
 # Train each model and calculate accuracies
train_accuracies = {}
test_accuracies = {}

for name, model in models.items():
    model.fit(XX_train, YY_train)  # Train model
    # Calculate accuracy
    train_accuracies[name] = accuracy_score(YY_train, model.predict(XX_train))
    test_accuracies[name] = accuracy_score(YY_test, model.predict(XX_test))
```

```python
 # Print the train accuracies
print("Training Data Accuracies:")
for name, acc in train_accuracies.items():
    print(f"{name}: {acc}")
 
 # Print the test accuarcies
print("\nTesting Data Accuracies:")
for name, acc in test_accuracies.items():
    print(f"{name}: {acc}")
```

```python
# Data for plotting
models = list(train_accuracies.keys())
training_acc = list(train_accuracies.values())
testing_acc = list(test_accuracies.values())

# Creating the bar plot
x = np.arange(len(models))  
width = 0.35  

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, training_acc, width, label='Training Accuracy')
rects2 = ax.bar(x + width/2, testing_acc, width, label='Testing Accuracy')

ax.set_ylabel('Accuracy')
ax.set_title('Accuracy by model and dataset')
ax.set_xticks(x)
ax.set_xticklabels(models, rotation=45, ha="right")
ax.legend()

fig.tight_layout()

plt.show()
```

# discusions
Logistic Regression, SVM with linear kernelVM with RBF kernel: These models performed well, likely because the decision boundaries between different Iris species can be approximated linearly with the chosen features.

Tree-Based Models: While powerful, the high training accuracy and lower testing accuracy of the Decision Tree and Random Forest models suggest they might be too complex for the relatively simple structure of the dataset, leading to overfitting.

KNN: The performance varied with the choice of k, emphasizing the importance of parameter tuning in KNN models. Smaller values of k led to slightly better generalization in this case.
